```
# File location and type
file_location = "/FileStore/tables/ParkingTotal2-1.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
%sql

select Registration_State,   count(Summons)  from `Data` group by Registration_State
ORDER BY count(Summons)ASC
```

```
# Create a view or table

temp_table_name = "Data"

df.createOrReplaceTempView(temp_table_name)
```

```
%sql

select  Sub_Division, count(Law_Section)   from `Data` group by Sub_Division 
 
```

```
%sql

select Violation_County,   count(Summons)  from `Data` group by Violation_County
ORDER BY count(Summons)ASC
```

```
%sql

select Sub_Division,  Violation_Description, Summons  from `Data`   group by Sub_Division
```

```
%sql

/* Query the created temp table in a SQL cell */

select Dia, count(Summons) from `Data` group by Dia
ORDER BY Dia ASC

```

```
%sql
select Mes,   count(Summons)  from `Data` group by Mes 
```

```
%sql

select Street_Name,   count(Summons)  from `Data` group by Street_Name
```
